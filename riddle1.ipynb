# Define the Riddle Puzzle Function
Define the riddle_puzzle function in a code cell.

```python
# Define the Riddle Puzzle Function
def riddle_puzzle():
    print("\nYou encounter a locked door with a note attached.")
    print("The note reads:")
    print("\n'Voiceless it cries,\nWingless flutters,\nToothless bites,\nMouthless mutters. What am I?'")

    correct_answer = "wind"

    while True:
        answer = input("\nWhat is your answer? ").strip().lower()
        if answer == correct_answer:
            print("\nYou hear the door opening.")
            print("You may now proceed.")
            break  # Exit the loop when the correct answer is given
        else:
            print("\nYou hear a loud click, but the door remains locked.")
            print("You must solve the riddle to proceed.")

```
